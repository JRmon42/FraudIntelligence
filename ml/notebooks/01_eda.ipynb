# 01 — EDA on synthetic Heimdall transactions

Quick exploratory look at the synthetic dataset used for offline training and demos. Run with: `jupyter nbconvert --to notebook --execute 01_eda.ipynb`.

```python
import sys, pathlib
sys.path.insert(0, str(pathlib.Path.cwd().parents[1]))
from ml.data.synthetic_data import SyntheticConfig, generate
import pandas as pd, numpy as np

df = generate(SyntheticConfig(n_transactions=20_000, seed=42))
print(df.shape, 'fraud_rate=', df['is_fraud'].mean().round(4))
df.head()
```

```python
df.groupby('card_country').agg(n=('tx_id','count'), fraud=('is_fraud','sum'), fraud_rate=('is_fraud','mean'))
```

```python
ax = df['amount'].clip(upper=1500).hist(bins=80, figsize=(8,3))
ax.set_title('Transaction amount distribution (clipped at 1500)')
```

```python
ring_summary = df[df['ring_id'] != ''].groupby('ring_id').agg(n_tx=('tx_id','count'), n_cards=('card_id','nunique'), n_devices=('device_id','nunique'))
ring_summary
```
